```python
#####################################################################################
class FunctionPlotter:
    def __init__(self, BW, PL):
        #initialise the instance of the button widgets class
        self.inst_bw = BW
        
        #initialise an instance of the plotting class
        self.inst_pl = PL
        
        #initialise the dropdown button
        #self.dropdown_button()

        #initialise widget to plot updaating graph in
        #self.plot_output = widgets.Output()

        #the box to hold upload button and dropdown button 
        first = widgets.HBox([self.inst_bw.upload, self.inst_pl.dropdown])
        
        #the box to hold the temp and pressure inputs
        first2 = widgets.HBox([self.inst_bw.temperature_box, self.inst_bw.pressure_box])
        
        #box to hold the updating plot output
        second = widgets.HBox([self.inst_pl.plot_output])
        
        #organising boxes into a left and right column
        left_column = widgets.VBox([first, first2, second])
        right_column = widgets.VBox([first2, self.inst_pl.plot2_output])
        
        #initialise the plot such that it shows immediatley
        #self.update_plot(self.dropdown.value)
        
        #display final interface
        display_layout = widgets.HBox([left_column, right_column])
        display(display_layout)

#####################################################################################
class ButtWidg:
    def __init__(self):
        self.input_params()
        self.upload_button()
        
    def input_params(self):
        default_temperature = '373'
        default_pressure = '100000'
        self.temperature_box = widgets.Text(value=default_temperature, layout={'width': '150px'}, description='Temp (K)')
        self.pressure_box = widgets.Text(value=default_pressure, layout={'width': '150px'}, description='Pressure (Pa)')
        
    def upload_button(self):
        self.upload = FileUpload(accept='.csv', multiple=True)

#####################################################################################
class Plots:
    def __init__(self):
        self.plot2_output = widgets.Output()
        self.plot_output = widgets.Output()
        self.plot2()
        self.dropdown_button()
        self.update_plot(self.dropdown.value)
        
    def plot2(self):
        with self.plot2_output:
            fig,ax = plt.subplots()
            ax.set(xlim=(-0.5,0.5), ylim=(-0.5,0.5))
            ax.axis('equal')

            centre_x = 0
            centre_y = 0
            ax.add_patch(plt.Circle((centre_x,centre_y),0.3, color='green'))
            ax.add_patch(plt.Circle((centre_x,centre_y),0.08, color='blue'))
            ax.annotate("Ro", xy=(centre_x, centre_y), xytext=(0.3+0.011, -0.01),arrowprops=dict(arrowstyle="]->", color='yellow'))
            ax.annotate("Rc", xy=(centre_x, centre_y), xytext=(-0.02, 0.08+0.02),arrowprops=dict(arrowstyle="]->", color='yellow'))
            plt.show()
            
    def dropdown_button(self):
        # Create a dropdown widget with options
        self.current_directory = os.getcwd()
        self.dropdown = widgets.Dropdown(
            options=self.get_file_list(self.current_directory+'/stakebake_data'),
            value = 'BM90.csv')
        # Set up the observe event
        self.dropdown.observe(self.update_plot, names='value')
    
    def get_file_list(self, folder_path):
        files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        return files

    def update_plot(self, change):
        if type(change)==str:
            selected_function = change
        else:
            selected_function = change['new']
        self.plot_output.clear_output(wait=True)
        data = pd.read_csv('stakebake_data/'+selected_function)
        with self.plot_output:
            x = data.x
            y=data[' y']

            plt.plot(x, y, label=selected_function)
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(f'Plot of {selected_function}')
            plt.legend()
            plt.show()
```
